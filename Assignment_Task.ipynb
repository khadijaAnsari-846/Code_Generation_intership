Task 1: String Analyzer Function

```python
#addition function needed


def isalpha(c):
  if(c>='a' and c<='z') or (c>='A' and c<='Z'):
    return True
  else:
    return False

def isnum(c):
  try:
    num=int(c)
    if(num>=0 and num<=9):
      return True
    else:
      return False
  except:
    return False
def isspace(c):
  if(c==' '):
    return True
  else:
    return False

def lower(c):   #convert into lower case
  if(c>='A' and c<='Z'):
    return chr(ord(c)+32)
  else:
    return c

def upper(c): #convert into upper case
  if(c>='a' and c<='z'):
    return chr(ord(c)-32)
  else:
    return c



# Convert string to int manually
def str_to_int(s):
  num = 0
  i = 0
  while i < len(s):
    num = num * 10 + (ord(s[i]) - ord('0'))
    i += 1
  return num




def float_to_str(n):
    # Convert integer part
    int_part = int(n)
    s = ""
    if int_part == 0:
        s = "0"
    else:
        digits = ""
        temp = int_part
        while temp > 0:
            digit = temp % 10
            digits = chr(ord('0') + digit) + digits
            temp = temp // 10
        s = digits

    # Simulate rounding
    frac = n - int_part
    frac_value = int(frac * 100 + 0.5)

    # Add decimal part
    s += "."

    if frac_value < 10:
        s += "0" + chr(ord('0') + frac_value)
    else:
        tens = frac_value // 10
        ones = frac_value % 10
        s += chr(ord('0') + tens) + chr(ord('0') + ones)

    return s


# Compare strings alphabetically: return True if a < b
def is_smaller(a, b):
  i = 0
  while i < len(a) and i < len(b):
    if a[i] < b[i]:
      return True
    elif a[i] > b[b]: # Fix typo: changed b[b] to b[i]
      return False
    i += 1
  if len(a) < len(b):
    return True
  return False
```

```python
def analyze_text(text):
  """
   ret: dict
   count: word, sentence count, vowel, digits
  """
  word_count = 0
  sentence_count = 0
  vowel_count = 0
  digit_count = 0
  is_word = False #just to maintain a record of words

  for letter in text:
    if isalpha(letter):
      vowel_count += 1 if lower(letter) in 'aeiou' else 0
      if not is_word:
        word_count += 1
        is_word = True
    elif isnum(letter):
      digit_count += 1
      if not is_word:
        word_count += 1
        is_word = True

    elif letter in ['.', '!', '?']:
      sentence_count += 1
      if not is_word:
        word_count += 1
        is_word = True
    elif isspace(letter):
      is_word = False
    else: # Handle other characters that are not alphanumeric, punctuation, or space
      is_word = True


  dict_counts = {
      "word_count": word_count,
      "sentence_count": sentence_count,
      "vowel_count": vowel_count,
      "digit_count": digit_count
  }

  return dict_counts



test_cases = [
    # Basic tests
    ("Hello world! How are you today?",
     {'word_count': 6, 'sentence_count': 2, 'vowel_count': 10, 'digit_count': 0}),

    ("",
     {'word_count': 0, 'sentence_count': 0, 'vowel_count': 0, 'digit_count': 0}),

    ("I have 5 apples and 3 oranges.",
     {'word_count': 7, 'sentence_count': 1, 'vowel_count': 9, 'digit_count': 2}),

    ("What?! Amazing!",
     {'word_count': 2, 'sentence_count': 3, 'vowel_count': 4, 'digit_count': 0}),

    ("A",
     {'word_count': 1, 'sentence_count': 0, 'vowel_count': 1, 'digit_count': 0}),

    # Edge cases
    ("123 456 789",
     {'word_count': 3, 'sentence_count': 0, 'vowel_count': 0, 'digit_count': 9}),

    ("   spaced   out   words   ",
     {'word_count': 3, 'sentence_count': 0, 'vowel_count': 5, 'digit_count': 0}),

    ("UPPERCASE lowercase MiXeD",
     {'word_count': 3, 'sentence_count': 0, 'vowel_count': 10, 'digit_count': 0}),

    ("No vowels? Try: bcdfg!",
     {'word_count': 4, 'sentence_count': 2, 'vowel_count': 3, 'digit_count': 0}),

    ("Multiple...periods...here.",
     {'word_count': 1, 'sentence_count': 7, 'vowel_count': 8, 'digit_count': 0}),
]



i = 0
while i < len(test_cases):
    data, expected = test_cases[i]
    result = analyze_text(data)
    print("Test Case", i + 1, "PASSED" if result == expected else "FAILED")
    print(result)
    print("---")
    i += 1

```

Task 2 : Feedback Classifier

```python
"""
simple logic is while iterating , see if any word in good and bad list match according to it count maintain and at end comapre a count ,
which one is greater return expression base on it.
 """
def classify_feedback(feedback):

  #first declare good and bad words
  postive_word=[ "good", "great", "excellent", "amazing", "love", "best", "wonderful", "fantastic" ]
  negative_word=["bad", "terrible", "awful", "hate", "worst", "horrible", "disappointing", "poor" ]



  if(feedback):
    pos_count=0
    neg_count=0
    letters=""

    for i in range(len(feedback)):
      if feedback[i]!=' ':
        letters+=lower(feedback[i])  #used lower func as per above declare to handle case-sensitivity

        if(i==len(feedback)-1):
          word=letters
          if word in postive_word:
            pos_count+=1
          elif word in negative_word:
            neg_count+=1
          letters=""
          word=""
      else:
        word=letters
        if word in postive_word:
          pos_count+=1
        elif word in negative_word:
          neg_count+=1
        letters=""
        word=""

    if(pos_count>neg_count):
      return "POSITIVE"
    elif(neg_count>pos_count):
      return "NEGATIVE"
    else:
      return "NEUTRAL"

  else:
    return "NEUTRAL"



#test case

test_cases = [
    # Basic positive
    ("This is good work", "POSITIVE"),
    ("Great job! Amazing results", "POSITIVE"),
    ("I love this excellent service", "POSITIVE"),

    # Basic negative
    ("This is bad work", "NEGATIVE"),
    ("Terrible service, awful experience", "NEGATIVE"),
    ("I hate this horrible product", "NEGATIVE"),

    # Neutral cases
    ("This is work", "NEUTRAL"),
    ("Good and bad mixed", "NEUTRAL"),
    ("", "NEUTRAL"),
    ("No sentiment words here", "NEUTRAL"),

    # Edge cases
    ("GOOD EXCELLENT", "POSITIVE"),             # Case insensitive
    ("good bad good", "POSITIVE"),              # More positive than negative
    ("bad terrible bad good", "NEGATIVE"),      # More negative than positive
    ("wonderful terrible", "NEUTRAL"),          # Equal count
]


i = 0
while i < len(test_cases):
    data, expected = test_cases[i]
    result = classify_feedback(data)
    print("Test Case", i + 1, "PASSED" if result == expected else "FAILED")
    print(result)
    print("---")
    i += 1



```

```python
# input_format:"ID:NAME:DEPARTMENT:SALARY"
"""
  as per the input format , if we exclude : than only having 4 word is valid.
  let just create a respective varible and validate it
"""
def validate_employee_data(data_string):


  id_flag=False # as we encounter a first : we store it in id_flag , like wise other
  name_flag=False
  department_flag=False
  salary_flag=False #special case where we did not seen any : so handle it separately

  id=""
  name=""
  department=""
  salary=""
  word=""


  #if nothing in string return false
  if(data_string):
    for i in range(len(data_string)):

      if data_string[i]==':' or i==len(data_string)-1:
        if id_flag==False:
          id=word
          id_flag=True
          word=""
        elif name_flag==False:
          name=word
          name_flag=True
          word=""
        elif department_flag==False:
          department=word
          department_flag=True
          word=""
        elif salary_flag==False:
          salary=word + data_string[i] #last elemt also need to be add
          salary_flag=True
          word=""
      else:
       word+=(data_string[i])



    #condition checking  -- for id
    for i in range(len(id)):
      if isnum(id[i]):
        id_flag=True
        if(i==len(id)-1):
          if(len(id)==6):
            id_flag=True
          else:
            id_flag=False
            break
      else:
        id_flag=False
        break




    #-- for name
    if(len(name)>0):
      for i in range(len(name)):
        if(isalpha(name[i]) ) or (isspace(name[i])):
          name_flag=True
        else:
          name_flag=False
          break
    else:
      name_flag=False



    #for department
    dep_valid=["HR", "IT", "FINANCE", "MARKETING", "OPERATIONS" ]

    if department in dep_valid:
      department_flag=True
    else:
      department_flag=False

    #for salary

    is_point=False
    roundoff_digit=""
    for i in range(len(salary)):
      if isnum(salary[i]) or (salary[i]=='.' and is_point==False): #to make sure that only one point allowed
        salary_flag=True


        #another condition that say len of salary after point(.) needed to be 2 only, also prohibit a second point
        if(salary[i]=='.'):
          is_point=True

        elif(is_point==True):
          roundoff_digit+=salary[i]


      else:
        salary_flag=False
        break



    if( salary_flag== True and is_point==True and len(roundoff_digit)==2):
      salary_flag=True


    if(id_flag==True and name_flag==True and department_flag==True and salary_flag==True):
      return True
    else:
     return False



  else:
    return False


test_cases = [
    # Valid cases
    ("123456:John Doe:IT:75000.50", True),
    ("789012:Jane Smith:HR:65000", True),
    ("111111:Bob Johnson:FINANCE:90000.00", True),
    ("222222:Alice Brown:MARKETING:55000", True),
    ("333333:Charlie Wilson:OPERATIONS:80000.25", True),

    # Invalid ID cases
    ("12345:John Doe:IT:75000", False),          # Too short
    ("1234567:John Doe:IT:75000", False),        # Too long
    ("12345A:John Doe:IT:75000", False),         # Contains letter

    # Invalid name cases
    ("123456:John123:IT:75000", False),          # Contains digit
    ("123456:John@Doe:IT:75000", False),         # Contains symbol
    ("123456::IT:75000", False),                 # Empty name

    # Invalid department cases
    ("123456:John Doe:Engineering:75000", False), # Invalid dept
    ("123456:John Doe:it:75000", False),         # Wrong case
    ("123456:John Doe::75000", False),           # Empty dept

    # Invalid salary cases
    ("123456:John Doe:IT:75000.50.25", False),   # Multiple decimals
    ("123456:John Doe:IT:75000A", False),        # Contains letter
    ("123456:John Doe:IT:", False),              # Empty salary
    ("123456:John Doe:IT:-5000", False),         # Negative (contains -)

    # Format issues
    ("123456:John Doe:IT", False),               # Missing field
    ("123456:John Doe:IT:75000:", False),        # Extra colon
    ("", False),                                 # Empty string
]




i = 0
while i < len(test_cases):
    data, expected = test_cases[i]
    result = validate_employee_data(data)
    print("Test Case", i + 1, "PASSED" if result == expected else "FAILED")
    print(result)
    print("---")
    i += 1
```

```python
def generate_summary_report(employee_records):
    total_employees = len(employee_records)

    if total_employees == 0:
        return "Total Employees: 0\nAverage Salary: 0.00\nLargest Department: None\nHighest Paid Department: None"

    total_salary = 0
    dept_counts = {}
    dept_salaries = {}

    #To extract manually (ID:Name:Dept:Salary)
    def parse_record(record):
        fields = []
        current = ""
        i = 0
        while i < len(record):
            if record[i] == ':':
                fields.append(current)
                current = ""
            else:
                current += record[i]
            i += 1
        fields.append(current)
        return fields  # [id, name, dept, salary]



    i = 0
    while i < total_employees:
        fields = parse_record(employee_records[i])
        dept = fields[2]
        salary = str_to_int(fields[3])
        total_salary += salary

        # update department count
        if dept in dept_counts:
            dept_counts[dept] += 1
            dept_salaries[dept] += salary
        else:
            dept_counts[dept] = 1
            dept_salaries[dept] = salary
        i += 1

    avg_salary = total_salary * 1.0 / total_employees

    # Find largest departments
    max_count = 0
    dept_list = []
    for dept in dept_counts:
        count = dept_counts[dept]
        if count > max_count:
            max_count = count
            dept_list = [dept]
        elif count == max_count:
            # insert in alphabetical order manually
            inserted = False
            j = 0
            while j < len(dept_list):
                if is_smaller(dept, dept_list[j]):
                    k = len(dept_list)
                    dept_list.append("")
                    while k > j:
                        dept_list[k] = dept_list[k - 1]
                        k -= 1
                    dept_list[j] = dept
                    inserted = True
                    break
                j += 1
            if not inserted:
                dept_list.append(dept)

    if len(dept_list) == 1:
        largest = dept_list[0]
    else:
        largest = ""
        j = 0
        while j < len(dept_list):
            largest += dept_list[j]
            if j != len(dept_list) - 1:
                largest += ", "
            j += 1
        largest += " (tie)"

    # Find highest average salary dept
    max_avg = -1
    highest_depts = []
    for dept in dept_salaries:
        avg = dept_salaries[dept] * 1.0 / dept_counts[dept]
        if avg > max_avg:
            max_avg = avg
            highest_depts = [dept]
        elif avg == max_avg:
            highest_depts.append(dept)

    # Find alphabetically first dept
    best_dept = highest_depts[0]
    j = 1
    while j < len(highest_depts):
        if is_smaller(highest_depts[j], best_dept):
            best_dept = highest_depts[j]
        j += 1

    return "Total Employees: " + str(total_employees) + \
           "\nAverage Salary: " + float_to_str(avg_salary) + \
           "\nLargest Department: " + largest + \
           "\nHighest Paid Department: " + best_dept






#test case
    test_cases = [
        # Basic case
        ([
             "123456:John Doe:IT:75000",
             "789012:Jane Smith:HR:65000"
         ],
         "Total Employees: 2\nAverage Salary: 70000.00\nLargest Department: HR, IT (tie)\nHighest Paid Department: IT"),

        # Single employee
        ([
             "123456:John Doe:IT:75000"
         ], "Total Employees: 1\nAverage Salary: 75000.00\nLargest Department: IT\nHighest Paid Department: IT"),

        # Multiple departments
        ([
             "123456:John Doe:IT:80000",
             "789012:Jane Smith:IT:70000",
             "111111:Bob Johnson:HR:60000",
             "222222:Alice Brown:FINANCE:90000"
         ], "Total Employees: 4\nAverage Salary: 75000.00\nLargest Department: IT\nHighest Paid Department: FINANCE"),

        # Tie in department size
        ([
             "123456:John Doe:IT:80000",
             "789012:Jane Smith:HR:60000",
             "111111:Bob Johnson:FINANCE:70000"
         ],
         "Total Employees: 3\nAverage Salary: 70000.00\nLargest Department: FINANCE, HR, IT (tie)\nHighest Paid Department: IT"),

        # Empty list
        ([], "Total Employees: 0\nAverage Salary: 0.00\nLargest Department: None\nHighest Paid Department: None"),
    ]

i = 0
while i < len(test_cases):
    data, expected = test_cases[i]
    result = generate_summary_report(data)
    print("Test Case", i + 1, "PASSED" if result == expected else "FAILED")
    print(result)
    print("---")
    i += 1

```
